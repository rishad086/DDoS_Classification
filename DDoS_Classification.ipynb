```python
# Apply IAT calculation for all datasets
for name, data in {
    "benign_train": benign_train,
    "dos_train": dos_train,
    "benign_test": benign_test,
    "dos_test": dos_test
}.items():
    # Sort by Time just in case
    data = data.sort_values('Time').reset_index(drop=True)
    data['IAT'] = data['Time'].diff().fillna(0)
    # Update the reference
    if name == "benign_train": benign_train = data
    elif name == "dos_train": dos_train = data
    elif name == "benign_test": benign_test = data
    elif name == "dos_test": dos_test = data
    print(f"âœ… Added IAT feature to {name} (shape: {data.shape})")
```

```python
# Apply rolling window features for all datasets
for name, data in {
    "benign_train": benign_train,
    "dos_train": dos_train,
    "benign_test": benign_test,
    "dos_test": dos_test
}.items():
    # Sort by time to ensure proper sequence
    data = data.sort_values('Time').reset_index(drop=True)
    
    # Rolling statistics over last 10 packets
    data['len_mean_10'] = data['Length'].rolling(window=10, min_periods=1).mean()
    data['len_std_10']  = data['Length'].rolling(window=10, min_periods=1).std().fillna(0)
    data['iat_mean_10'] = data['IAT'].rolling(window=10, min_periods=1).mean()
    
    # Re-assign back to variable
    if name == "benign_train": benign_train = data
    elif name == "dos_train": dos_train = data
    elif name == "benign_test": benign_test = data
    elif name == "dos_test": dos_test = data
    
    print(f"âœ… Added rolling features to {name} (shape: {data.shape})")
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import mutual_info_classif
from sklearn.ensemble import RandomForestClassifier

# ----- Prepare features and labels -----


# ---------- Fisher Score ----------
def fisher_score_binary(X, y):
    X = X.to_numpy().astype(float)
    y = y.to_numpy()
    scores = np.zeros(X.shape[1])
    mask1, mask0 = (y == 1), (y == 0)
    X1, X0 = X[mask1], X[mask0]
    for j in range(X.shape[1]):
        m1, m0 = X1[:, j].mean(), X0[:, j].mean()
        v1, v0 = X1[:, j].var(), X0[:, j].var()
        scores[j] = ((m1 - m0) ** 2) / (v1 + v0 + 1e-12)
    return scores

fisher_scores = fisher_score_binary(X, y)
X = train_df.drop(columns=['Attack']).to_numpy()
y = train_df['Attack'].to_numpy()
# ---------- Mutual Information ----------
mi_scores = mutual_info_classif(X, y, random_state=42)

# ---------- Random Forest Feature Importance ----------
rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X, y)
rf_scores = rf.feature_importances_

# ---------- Combine Results ----------
comparison = pd.DataFrame({
    'Feature': feature_names,
    'Fisher_Score': fisher_scores,
    'Mutual_Info': mi_scores,
    'RF_Importance': rf_scores
})

# Rank within each method
for col in ['Fisher_Score', 'Mutual_Info', 'RF_Importance']:
    comparison[f'{col}_Rank'] = comparison[col].rank(ascending=False)

# Compute average rank
comparison['Avg_Rank'] = comparison[[f'{c}_Rank' for c in ['Fisher_Score','Mutual_Info','RF_Importance']]].mean(axis=1)

# Sort by average rank
comparison = comparison.sort_values('Avg_Rank').reset_index(drop=True)

print("ðŸ“Š Feature Importance Comparison Table")
display(comparison)
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import silhouette_score
import numpy as np

# 1ï¸âƒ£ Choose features for Isolation Forest (paper-style)
# For DoS in Bluetooth, they used top features (e.g. Time & Info).
# Adjust if your column names differ.
features_if = []

if 'Time' in train_df.columns:
    features_if.append('Time')
if 'Info' in train_df.columns:
    features_if.append('Info')

# Fallback: if for some reason 'Info' is not encoded, use another strong feature
if len(features_if) < 2:
    # e.g., add Length or IAT
    if 'Length' in train_df.columns:
        features_if.append('Length')
    elif 'IAT' in train_df.columns:
        features_if.append('IAT')

print("Using features for Isolation Forest:", features_if)

X_if = train_df[features_if].astype(float)

# 2ï¸âƒ£ Subsample 10,000 points (or less if dataset smaller)
sample_size = min(10000, len(X_if))
X_sample = X_if.sample(n=sample_size, random_state=42)

print("Sampled size for IF + silhouette:", X_sample.shape[0])

# 3ï¸âƒ£ Train & evaluate two Isolation Forest models
for contamination in [0.1, 0.3]:
    if_model = IsolationForest(
        n_estimators=200,
        contamination=contamination,
        random_state=42,
        n_jobs=-1
    )

    # Fit on the sample (unsupervised)
    if_model.fit(X_sample)

    # Predict: -1 = anomaly, 1 = normal
    y_pred_if = if_model.predict(X_sample)

    # Convert to cluster labels: 0 = normal, 1 = anomaly
    clusters = (y_pred_if == -1).astype(int)

    # Need at least 2 clusters for silhouette_score
    if len(np.unique(clusters)) < 2:
        print(f"\n===== IF (contamination={contamination}) =====")
        print("Silhouette Score: cannot compute (only one cluster found).")
        continue

    sil_score = silhouette_score(X_sample, clusters)

    print(f"\n===== Isolation Forest (contamination={contamination}) =====")
    print(f"Silhouette Score: {sil_score:.4f}")
```
